```python
import json

with open('/Users/kaizhang/Documents/0xplain/0xplain/datastory/data/ekubo.json', 'r') as file:
    data = json.load(file)
```

```python
# Extracting the TVL data for "Starknet-borrowed"
tvl_data = data['chainTvls']['Starknet']['tvl']

# Extracting dates and TVL values
dates = [item['date'] for item in tvl_data]
values = [item['totalLiquidityUSD'] for item in tvl_data]
```

```python
import pandas as pd

# Assuming `dates` and `values` are your lists of dates and TVLs
data = {'Date': dates, 'TVL': values}
df = pd.DataFrame(data)
```

```python
import plotly.express as px
import pandas as pd

# Your DataFrame creation here
df = pd.DataFrame({
    'Date': pd.to_datetime(dates),  # Converts to datetime if not already
    'TVL': values,
})


# Create the line chart
fig = px.line(df, x='Date', y='TVL', title='Total Value Locked (TVL) Over Time')

# Customize the chart for a cleaner look
fig.update_layout({
    'plot_bgcolor': 'rgba(0,0,0,0)',
    'xaxis_title': 'Date',
    'yaxis_title': 'Total Value Locked (USD)',
    'xaxis': {
        'showline': True,
        'showgrid': False,
        'linecolor': 'gray',
    },
    'yaxis': {
        'showgrid': True,
        'gridcolor': 'lightgray',
    },
})

# # Add hover tool for interactivity
fig.update_traces(mode='lines+markers', hoverinfo='x+y', marker=dict(size=5))


# Simplified annotation addition
# Ensure annotation_date is within the df['Date'] range
annotations = [
    ('2024-02-1T17:00:00.000000000', "News came out"),
    ('2024-02-10T17:00:00.000000000', "Airdrop TGE"),
    # Add more dates and events as needed
]

for date_str, text in annotations:
    annotation_date = pd.to_datetime(date_str)
    if annotation_date in df['Date'].values:
        tvl_value = df[df['Date'] == annotation_date]['TVL'].iloc[0]
        fig.add_annotation(x=annotation_date, y=tvl_value,
                           text=text,
                           showarrow=True,
                           arrowhead=1,
                           ax=0,  # Adjust if needed
                           ay=-40,  # Adjust if needed
                           bgcolor="lightyellow",
                           font=dict(size=12))
else:
   print("Date not found in DataFrame.")

# Show the plots
fig.show()
```
